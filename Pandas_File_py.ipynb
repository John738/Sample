```python
def get_customer_orders(orders, customer_id):
    orders_filtered = []
    for order in orders:
        if int(order.split(',')[2]) == customer_id:
            orders_filtered.append(order)
    return orders_filtered
```

```python
def get_order_details(orders,cust_id):
    for order in orders:
        if (order.split(',')[2]) == cust_id:
            print (order)
```
